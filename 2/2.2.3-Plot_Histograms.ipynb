```python
import os
import numpy as np
import rasterio as rio
import matplotlib.pyplot as plt
import seaborn as sns

import earthpy as et
import earthpy.plot as ep

# Prettier plotting with seaborn
sns.set(font_scale=1.5, style="whitegrid")
```

```python
with rio.open('/home/marta/PycharmProjects/Earth analytics/spatial//boulder-leehill-rd/pre-flood/lidar/pre_DTM.tif') as src:
    lidar_dem_im = src.read(1, masked=True)

# View object dimensions
lidar_dem_im.shape
```

```python
print(lidar_dem_im.shape)
print(lidar_dem_im.ravel().shape)
```

```python
lidar_dem_hist = lidar_dem_im.ravel()
lidar_dem_hist[~np.isnan(lidar_dem_hist)]
```

```python
print(lidar_dem_hist.shape)
```

```python
ep.hist(lidar_dem_im, colors=['purple'],
        title="Distribution of Lidar DEM Elevation Values",
        xlabel='Elevation (meters)',
        ylabel='Frequency')
plt.show()
```

```python
ep.hist(lidar_dem_im, colors=['purple'],
        title="Distribution of Lidar DEM Elevation Values",
        xlabel='Elevation (meters)',
        ylabel='Frequency',
        bins=7)
plt.show()

```

```python
ep.hist(lidar_dem_im,
        bins=[1600, 1800, 2000, 2100],
        colors=['purple'],
        title="Distribution of Lidar DEM Elevation Values",
        xlabel='Elevation (meters)',
        ylabel='Frequency')
plt.show()
```
