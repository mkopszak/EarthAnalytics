```python
import os
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
import earthpy as et
import earthpy.plot as ep
```

```python
with rio.open("/home/marta/PycharmProjects/Earth analytics/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM.tif") as lidar_dem:
    lidar_dem_im = lidar_dem.read(1, masked=True)

# Open dem hillshade
with rio.open("/home/marta/PycharmProjects/Earth analytics/spatial/boulder-leehill-rd/pre-flood/lidar/pre_DTM_hill.tif") as lidar_dem_hill:
    lidar_dem_hill = lidar_dem_hill.read(1, masked=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ep.plot_bands(lidar_dem_im, ax=ax, cmap='viridis_r',
              title="Lidar Digital Elevation Model (DEM)\n overlayed on top of a hillshade")
ax.imshow(lidar_dem_hill, cmap='Greys', alpha=.5)
ax.set_axis_off()
plt.show()
```
